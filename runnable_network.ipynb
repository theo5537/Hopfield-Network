```python

# Training Hopfield Network
def train_hopfield(images):
    num_patterns, num_neurons = images.shape

    W = np.zeros((num_neurons, num_neurons))

    for pattern in images:
        W += np.outer(pattern, pattern)

    np.fill_diagonal(W, 0)

    return W

W = train_hopfield(train_images)
print("Hopfield Network Trained! Weight matrix shape:", W.shape)

# Update Functions
def asynchronous_update(W, pattern, steps=2000):
   #Update one random neuron at a time
    num_neurons = pattern.shape[0]
    pattern = pattern.copy()

    for _ in range(steps):
        i = np.random.randint(num_neurons)
        pattern[i] = np.sign(W[i] @ pattern)
        if pattern[i] == 0:
            pattern[i] = 1

    return pattern

def synchronous_update(W, pattern, max_steps=100):
    #Update all neurons at once
    pattern = pattern.copy()  # Avoid modifying the input pattern
    for step in range(1, max_steps + 1):
        new_pattern = np.sign(W @ pattern)
        new_pattern[new_pattern == 0] = 1
# Check convergence
        if np.array_equal(new_pattern, pattern):
            return new_pattern, step

        pattern = new_pattern  # Update pattern

    return pattern, max_steps


def corrupt_pattern(pattern, corruption_prob):
   # Flips each pixel with probability p
    corrupt_mask = np.random.rand(pattern.shape[0]) < corruption_prob
    corrupted_pattern = pattern.copy()
    corrupted_pattern[corrupt_mask] *= -1
    return corrupted_pattern
def crop(pattern, img_size=16, box_size=10):
    pattern = pattern.reshape(img_size, img_size).copy()

    # Define bounding box start and end indices
    start = (img_size - box_size) // 2
    end = start + box_size

    # Create a mask to keep only the bounding box unchanged
    mask = np.zeros((img_size, img_size), dtype=bool)
    mask[start:end, start:end] = True  # Set center to True

    # Choose all black or all white outside the box
    outside_value = np.random.choice([-1, 1])
    pattern[~mask] = outside_value

    return pattern.flatten()

# Corrupt and Recover Patterns
#Set probability here
corruption_prob = 0.1
corrupted_patterns = np.array([corrupt_pattern(img, corruption_prob) for img in train_images])

# Store recovered patterns and step counts separately
recovered_sync_results = [synchronous_update(W, img) for img in corrupted_patterns]

# Extract final patterns and steps taken
recovered_sync_patterns = np.array([result[0] for result in recovered_sync_results])  # Extract only patterns
sync_convergence_steps = np.array([result[1] for result in recovered_sync_results])  # Extract step counts

recovered_async_patterns = np.array([asynchronous_update(W, img) for img in corrupted_patterns])

# Reshape for visualization
originals = train_images.reshape(num_train_images, 16, 16)
corrupteds = corrupted_patterns.reshape(num_train_images, 16, 16)
sync_recovereds = recovered_sync_patterns.reshape(num_train_images, 16, 16)
async_recovereds = recovered_async_patterns.reshape(num_train_images, 16, 16)

# Plot results
def visualize_all_patterns(originals, corrupteds, sync_recovereds, async_recovereds, sync_steps, cols=4):
    num_images = originals.shape[0]
    rows = (num_images + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols * 4, figsize=(cols * 4, rows * 4))

    for i in range(num_images):
        row, col_offset = divmod(i, cols)
        col_start = col_offset * 4

        axes[row, col_start].imshow(originals[i], cmap="gray")
        axes[row, col_start].set_title(f"Original {i}", fontsize=8)
        axes[row, col_start].axis("off")

        axes[row, col_start + 1].imshow(corrupteds[i], cmap="gray")
        axes[row, col_start + 1].set_title("Corrupted", fontsize=8)
        axes[row, col_start + 1].axis("off")

        axes[row, col_start + 2].imshow(sync_recovereds[i], cmap="gray")
        axes[row, col_start + 2].set_title(f"Sync Recovered\n(Steps: {sync_steps[i]})", fontsize=8)
        axes[row, col_start + 2].axis("off")

        axes[row, col_start + 3].imshow(async_recovereds[i], cmap="gray")
        axes[row, col_start + 3].set_title("Async Recovered", fontsize=8)
        axes[row, col_start + 3].axis("off")

    plt.tight_layout()
    plt.show()


visualize_all_patterns(originals, corrupteds, sync_recovereds, async_recovereds, sync_convergence_steps)
```

```python
def track_synchronous_update(W, pattern, max_steps=100):
    history = [pattern.copy()]
    for _ in range(max_steps):
        new_pattern = np.sign(W @ history[-1])
        new_pattern[new_pattern == 0] = 1  # Ensure only {-1,1}
        if np.all(new_pattern == history[-1]):
            break  # Converged
        history.append(new_pattern.copy())  # Store intermediate step
    return history
def track_asynchronous_update(W, pattern, max_steps=2000, step_interval=100):
    num_neurons = pattern.shape[0]
    history = [pattern.copy()]
    current_pattern = pattern.copy()  # Maintain a separate mutable copy

    for step in range(1, max_steps + 1):
        i = np.random.randint(num_neurons)  # Pick a random neuron
        new_value = np.sign(W[i] @ current_pattern)
        new_value = 1 if new_value == 0 else new_value  # Ensure {-1,1}

        if new_value != current_pattern[i]:  # Update only if there's a change
            current_pattern[i] = new_value

        # Store pattern every `step_interval` updates
        if step % step_interval == 0 or step == max_steps:
            history.append(current_pattern.copy())

        # Convergence check: Stop early if the last two patterns are identical
        if len(history) > 1 and np.all(history[-1] == history[-2]):
            break

    # Ensure the final converged pattern is stored
    history.append(current_pattern.copy())
    return history


def visualize_recovery_process(history, title="Recovery Process"):
    steps = len(history)
    fig, axes = plt.subplots(1, steps, figsize=(steps * 2, 2))
    for i, pattern in enumerate(history):
        axes[i].imshow(pattern.reshape(16, 16), cmap="gray")
        axes[i].axis("off")
        axes[i].set_title(f"Step {i}")
    plt.suptitle(title)
    plt.show()
```
